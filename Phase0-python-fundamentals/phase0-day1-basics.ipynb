```python
#Strings
s = "abcdefghi"
print(s[0]) #index 0
print(s[5]) #index 5
```

```python
#Slicing
print(s[0:4]) #first 4 letters (index 0 to 3)
print(s[2:7]) #index 2 to index 6
print(s[:5]) #index 0 to 4
print(s[3:]) #index 3 to last
```

```python
#stringMethods
name = "ShAshWat"
print(name.lower()) #lowerCases
print(name.upper()) #upperCases

s = " hello world "
print(s.strip()) #trim whitespaces

print("I hate python".replace("hate", "love")) #replaceString
print("banana".count("a")) #count orrucence

print("hello".startswith("he")) #checkStart
print("hello".endswith("lo")) #checkend
```
